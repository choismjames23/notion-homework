__[문제 2: 붕어빵 주문 받기]__

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
if 재고[종류] >= 주문[종류]:
    재고[종류] -= 주문[종류]
    print(f'주문하신 {종류}붕어빵 {주문[종류]}개 입니다~!')
else:
    print(f'주문하신 {종류}붕어빵은 재고가 {재고[종류]}개로 충분하지 않습니다.')    
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
재고 = {
    "팥" : 10,
    "슈크림" : 8,
    "초코" : 5
}
주문 = {}

while True:
    종류 = input("주문할 붕어빵 종류를 입력하세요(팥, 슈크림, 초코)(종료 입력시 주문 완료): ")
    if 종류 == "종료":
        break
    
    개수 = int(input("주문할 붕어빵 개수를 입력하세요: "))
    주문[종류] = 개수

    if 재고[종류] >= 주문[종류]:
        재고[종류] -= 주문[종류]
        주문[종류] = 개수
        print(f'주문하신 {종류}붕어빵 {주문[종류]}개 입니다~!')
    else:
        print(f'주문하신 {종류}붕어빵은 재고가 {재고[종류]}개로 충분하지 않습니다.')

print("현재 붕어빵 재고 : ", end="")
print(재고)
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
재고 = {
    "팥" : 10,
    "슈크림" : 8,
    "초코" : 5
}
주문 = {}
재고추가 = {}

while True:
    종류 = input("주문할 붕어빵 종류를 입력하세요(팥, 슈크림, 초코)(종료 입력시 주문 완료): ")
    if 종류 == "종료":
        break
    
    elif 종류 == "관리자":
        while True:
            종류2 = input("재고에 추가할 붕어빵 종류를 입력하세요(팥, 슈크림, 초코): ")
            if 종류2 == "완료":
                break
            개수2 = int(input("추가할 붕어빵 개수를 입력하세요: "))
            재고추가[종류2] = 개수2
            재고[종류2] += 재고추가[종류2]
    
        continue

    개수 = int(input("주문할 붕어빵 개수를 입력하세요: "))
    주문[종류] = 개수

    if 재고[종류] >= 주문[종류]:
        재고[종류] -= 주문[종류]
        주문[종류] = 개수
        print(f'주문하신 {종류}붕어빵 {주문[종류]}개 입니다~!')
    else:
        print(f'주문하신 {종류}붕어빵은 재고가 {재고[종류]}개로 충분하지 않습니다.')

print("현재 붕어빵 재고 : ", end="")
print(재고)
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
재고 = {
    "팥" : 10,
    "슈크림" : 8,
    "초코" : 5
}
가격 = {
    "팥" : 1000,
    "슈크림" : 1200,
    "초코" : 1500
}
주문 = {}
재고추가 = {}
주문금액 = 0
while True:
    종류 = input("주문할 붕어빵 종류를 입력하세요(팥, 슈크림, 초코)(종료 입력시 주문 완료): ")
    if 종류 == "종료":
        break
    
    elif 종류 == "관리자":
        while True:
            종류2 = input("재고에 추가할 붕어빵 종류를 입력하세요(팥, 슈크림, 초코): ")
            if 종류2 == "완료":
                break
            개수2 = int(input("추가할 붕어빵 개수를 입력하세요: "))
            재고추가[종류2] = 개수2
            재고[종류2] += 재고추가[종류2]
            print(f'{종류2}붕어빵 {개수2}개 추가완료! 현재 재고: {재고[종류2]}개')
    
        continue

    개수 = int(input("주문할 붕어빵 개수를 입력하세요: "))
    주문[종류] = 개수

    if 재고[종류] >= 주문[종류]:
        재고[종류] -= 주문[종류]
        주문[종류] = 개수
        print(f'주문하신 {종류}붕어빵 {주문[종류]}개 입니다~!')
    else:
        print(f'주문하신 {종류}붕어빵은 재고가 {재고[종류]}개로 충분하지 않습니다.')

for 종류 in 주문:
    주문금액 += 주문[종류]*가격[종류]

print("현재 붕어빵 재고 : ", end="")
print(재고)
print(f'총 주문한 내용은 팥붕어빵 {주문["팥"]}개, 슈크림붕어빵 {주문["슈크림"]}개, 초코붕어빵 {주문["초코"]}개 입니다.')
print(f'전체 주문 금액은 {주문금액}원 입니다.')
```
